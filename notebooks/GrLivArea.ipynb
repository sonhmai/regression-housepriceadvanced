GrLivArea - above grade (ground) living area square feet
- log transformation to fix right-skewed distribution does not have effect on kaggle score. even decreased score for 0.00001 from 0.13674 to 0.13675

```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
PATH='../data/'
train = pd.read_csv(PATH + 'train.csv')
test = pd.read_csv(PATH + 'test.csv')

area_train = train['GrLivArea']
area_test = test['GrLivArea']
```

```python
sns.boxplot(area_train)
```

```python
sns.boxplot(area_test)
```

```python
area_train.hist()
```

```python
area_test.hist()
```

## Log transform
From above histogram, both living area from train and test set has right-skewed distributions => log transform might help.

=> naturally, they have log-normal distributions because there is a lot houses with small or normal area, while there are some special houses that are large.

```python
area_train.apply(np.log).hist()
```

```python
area_test.apply(np.log).hist()
```

## Remove outliers

```python
from scipy import stats

print(area_train.shape)
not_outlier_filter = np.abs(stats.zscore(area_train)) < 2.5
print(area_train[not_outlier_filter].shape)
print(train[not_outlier_filter].shape)
```
